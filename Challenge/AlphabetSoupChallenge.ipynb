```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import r2_score
import pandas as pd
import tensorflow as tf

# Import our input dataset
alphabet_df = pd.read_csv('charity_data.csv')
alphabet_df
```

```python
alphabet_df.dtypes
```

```python
alphabet_df.nunique()
```

```python
application_counts = alphabet_df.APPLICATION_TYPE.value_counts()
print(application_counts)
```

```python
# We need bucketing for APPLICATION_TYPE
# Determine which values need to be replace in APPLICATION_TYPE
replace_application = list(application_counts[application_counts<100].index)

# Replace in DataFrame
for application in replace_application:
    alphabet_df.APPLICATION_TYPE = alphabet_df.APPLICATION_TYPE.replace(application,"Other")

    
# Check to make sure binning was successful
alphabet_df.APPLICATION_TYPE.value_counts()
```

```python
classification_counts = alphabet_df.CLASSIFICATION.value_counts()
print(classification_counts)
```

```python
# We need bucketing for CLASSIFICATION
# Determine which values need to be replace in CLASSIFICATION
replace_classification = list(classification_counts[classification_counts<10].index)

# Replace in DataFrame
for classification in replace_classification:
    alphabet_df.CLASSIFICATION = alphabet_df.CLASSIFICATION.replace(classification,"Other")

    
# Check to make sure binning was successful
alphabet_df.CLASSIFICATION.value_counts().head(30)
```

```python
print(f"Checking for data types: {alphabet_df.dtypes}")
print(f"Checking for unique values count: {alphabet_df.nunique()}")
print(f"Checking for Null Values: {alphabet_df.isnull().sum()}")
```

```python
# Classify columns by numeric and categorical
numeric_columns = ['STATUS', 'ASK_AMT', 'IS_SUCCESSFUL']
category_columns = ['APPLICATION_TYPE','AFFILIATION', 'CLASSIFICATION', 'USE_CASE', 'ORGANIZATION',
                   'INCOME_AMT','SPECIAL_CONSIDERATIONS']
```

```python
# Create a OneHotEncoder instance
enc = OneHotEncoder(sparse=False)

# Fit and transform the OneHotEncoder using the categorical variable list
encode_df = pd.DataFrame(enc.fit_transform(alphabet_df[category_columns]))

# Add the encoded variable names to the DataFrame
encode_df.columns = enc.get_feature_names(category_columns)
encode_df.head()


df_numeric = alphabet_df[numeric_columns].merge(encode_df, left_index=True, right_index=True)
df_numeric
```

```python
# Remove IS_SUCCESSFUL target from features data
y = df_numeric.IS_SUCCESSFUL
X = df_numeric.drop(columns=['IS_SUCCESSFUL'])


# Split training/test datasets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

# Create a StandardScaler instance
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Define the model - deep neural net
number_input_features = len(X_train_scaled[0])
hidden_nodes_layer1 =  69
hidden_nodes_layer2 = 35


nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(
    tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="relu")
)

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="relu"))


# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Compile the Sequential model together and customize metrics
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])

# Summary
nn.summary()
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled, y_train, epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Running the Random Forest Classifier to check if this model does a better job than the Neural Network
from sklearn.ensemble import RandomForestClassifier
# Create a random forest classifier.
rf_model = RandomForestClassifier(n_estimators=180, random_state=78)

# Fitting the model
rf_model = rf_model.fit(X_train_scaled, y_train)

# Evaluate the model
y_pred = rf_model.predict(X_test_scaled)
print(f" Random forest predictive accuracy: {accuracy_score(y_test,y_pred):.3f}")
```

```python
# Running the Logistic Regression to check if this model does a better job than the Neural Network
from sklearn.linear_model import LogisticRegression
# Define the logistic regression model
log_classifier = LogisticRegression(solver="lbfgs",max_iter=200)

# Train the model
log_classifier.fit(X_train,y_train)

# Evaluate the model
y_pred = log_classifier.predict(X_test)
print(f" Logistic regression model accuracy: {accuracy_score(y_test,y_pred):.3f}")
```
